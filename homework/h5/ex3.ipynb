## Problem 2

```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from matplotlib.colors import ListedColormap


sc= SparkContext()
sqlContext = SQLContext(sc)
```

```python
sensor_1 = "./sensors1.csv"
sensor_1_df = sqlContext.read.format('csv').options(header='false').load(sensor_1)
df1 = sensor_1_df.toPandas()
df1.shape
```

```python
Y = df1.loc[:, df1.columns[-1]]
X = df1.drop([df1.columns[-1]], axis=1)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
df_pca = pca.fit_transform(X=X)
pca.explained_variance_ratio_.round(2)
```

## Problem 3

```python
from sklearn.linear_model import LinearRegression
X = df1.loc[:, df1.columns[1:4]]
lr = LinearRegression(fit_intercept=True).fit(X, Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X,Y)
```
